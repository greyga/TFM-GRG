```r
# descargar el archivo genes_to_phenotype.txt
system("wget http://purl.obolibrary.org/obo/hp/hpoa/genes_to_phenotype.txt")
```

```r
# Leer el archivo especificando que es un archivo de texto con separador de tabulaciones
genes_pheno <- read.delim("/content/genes_to_phenotype.txt", header = TRUE, sep = "\t")


# Función que busca los genes con un HPO y también con el HPO Autosomal recessive inheritance
find_genes <- function(hpo) {
  # Filtrar los genes con el HPO especificado
  genes_hpo <- genes_pheno[genes_pheno$hpo_id == hpo, "gene_symbol"]

  # Filtrar los genes con el HPO Autosomal recessive inheritance
  genes_ar <- genes_pheno[genes_pheno$hpo_id == "HP:0000007", "gene_symbol"]

  # Encontrar los genes que están en ambos grupos
  genes_result <- intersect(genes_hpo, genes_ar)

  # Devolver los resultados
  return(genes_result)
}

# Ejemplo de uso
find_genes("HP:0002099") # busca los genes con el HPO "HP:0002460" y también con el HPO "HP:0000007"
```

```r
# Función que busca los genes comunes a una lista de HPOs y muestra los HPOs asociados a cada gen
find_common_genes <- function(hpo_list) {
  # Filtrar los genes para cada HPO en la lista
  genes_list <- lapply(hpo_list, function(hpo) {
    genes_pheno[genes_pheno$hpo_id == hpo, "gene_symbol"]
  })

  # Encontrar los genes comunes a todos los conjuntos
  genes_result <- Reduce(intersect, genes_list)

  # Crear un data frame vacío para almacenar los resultados
  result_df <- data.frame(gene_symbol = character(),
                          hpo_id = character(),
                          num_hpos = integer())

  # Obtener los HPOs asociados a cada gen y agregarlos al data frame de resultados
  for (gene in genes_result) {
    hpos <- genes_pheno[genes_pheno$gene_symbol == gene, "hpo_id"]
    # Eliminar los HPOs de la lista original
    hpos <- hpos[!hpos %in% hpo_list]
    # Devolver los HPOs únicos
    hpos <- unique(hpos)
    # Agregar los resultados al data frame
    result_df <- rbind(result_df, data.frame(gene_symbol = gene,
                                             hpo_id = toString(hpos),
                                             num_hpos = length(hpos)))
  }

  # Devolver los resultados
  return(result_df)
}
```

```r
find_associated_hpos <- function(gene_list) {
  # Convertir la entrada en una lista si es un solo gen
  if (is.character(gene_list)) {
    gene_list <- list(gene_list)
  }

  # Filtrar los HPOs para cada gen en la lista
  hpos_list <- lapply(gene_list, function(gene) {
    genes_pheno[genes_pheno$gene_symbol == gene, "hpo_id"]
  })

  # Combinar los HPOs en un vector único
  hpos_result <- unlist(hpos_list)

  # Obtener los HPOs únicos y su frecuencia
  unique_hpos <- unique(hpos_result)
  hpo_freq <- table(hpos_result)

  # Crear un data frame con los resultados
  result_df <- data.frame(hpo_id = character(),
                          frequency = integer())

  # Agregar los resultados al data frame
  for (hpo in unique_hpos) {
    freq <- hpo_freq[hpo]
    result_df <- rbind(result_df, data.frame(hpo_id = hpo,
                                             frequency = freq))
  }

  # Ordenar los resultados por frecuencia descendente
  result_df <- result_df[order(-result_df$frequency), ]

  # Devolver los resultados
  return(result_df)
}
find_associated_hpos("PTGER2")
```

```r
find_gene_hpo_associations <- function(df, gene_symbols) {
  result_df <- data.frame(gene_symbol = character(),
                          hpo_id = character(),
                          total_hpos = integer(),
                          stringsAsFactors = FALSE)

  for (gene_symbol in gene_symbols) {
    # Filtrar el data frame por el gen dado
    gene_df <- df[df$gene_symbol == gene_symbol, ]

    # Obtener los HPOs asociados al gen
    hpo_ids <- unique(gene_df$hpo_id)
    total_hpos <- length(hpo_ids)

    # Crear un data frame de resultados para el gen actual
    gene_result <- data.frame(gene_symbol = gene_symbol,
                              hpo_id = toString(hpo_ids),
                              total_hpos = total_hpos,
                              stringsAsFactors = FALSE)

    # Agregar el resultado a la tabla final
    result_df <- rbind(result_df, gene_result)
  }

  # Restablecer los índices del data frame de resultados
  rownames(result_df) <- NULL

  # Devolver el data frame de resultados
  return(result_df)
}
```

```r
# usar la funcion con el data frame genes_pheno

library(dplyr)

gene_symbols <- c("PTGER2", "ADRB2", "HNMT", "TBX21", "PLA2G7", "SLC27A4", "CARD10", "RIC1", "GP9", "IL21R")
gene_hpo_associations <- find_gene_hpo_associations(genes_pheno, gene_symbols) %>%
  arrange(total_hpos)

gene_hpo_associations
```

```r
library(HPOSim)

calculate_hpo_similarity <- function(hpo_list_of_interest, gene) {
  # Obtener los HPO asociados al gen especificado
  hpos_gene <- genes_pheno[genes_pheno$gene_symbol == gene, "hpo_id"]

  # Calcular la similitud semántica entre los HPO de interés y los HPO asociados al gen
  similarity_scores <- getTermListSim(hpo_list_of_interest, hpos_gene)

  # Crear un data frame con los resultados de similitud
  result_df <- data.frame(hpo_id = hpos_gene,
                          similarity_score = similarity_scores)

  return(result_df)
}
```
